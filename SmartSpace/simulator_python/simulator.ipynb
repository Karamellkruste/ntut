```python
def add(data):
    timestamp = str(int(time.time() * 1000000))
    # 持續增加數據
    # ref = db.reference('datas/' + timestamp)
    # 數據更新
    ref = db.reference('datas/')
    ref.set(data)
    pprint('Send data to db: {}'.format(data))
```

```python
def create_data():
    global used_coords_count
    pm25 = random.randint(20, 40)
    pm100 = random.randint(15, 35)
    o3 = random.randint(80, 130)
    so2 = random.randint(60, 90)
    no2 = random.randint(100, 120)
    co = random.randint(30, 50)
    temperature = random.uniform(28.0, 30.0)
    humidity = random.uniform(60.0, 68.0)

    # 獲得現在時間
    now = datetime.datetime.now()
    now_timezone = now + datetime.timedelta(hours=8)
    currentTime = now_timezone.strftime("%Y/%m/%d %H:%M:%S")

    # 用於依序取出座標點數據，超出最大數目後從頭取
    if used_coords_count >= len(coords_data):
        used_coords_count = 0
    coord = coords_data[used_coords_count]
    used_coords_count += 1


    return {'pm25': pm25, 'pm100': pm100, 'o3': o3, 'so2': so2, 'no2': no2, 'co': co, 'temperature': temperature, 'humidity': humidity, 'time': currentTime, 'coords': coord}
```

```python
used_coords_count = 0

while True:
    add(create_data())
    time.sleep(1)

print('Done')
```
